```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
from random import shuffle
```

```python
#从xlsx文件读取到pandas的dataframe结构中
df = pd.read_excel('1.xlsx')
```

```python
from sklearn.utils import shuffle
#对数据进行shuffle操作，打乱顺序
df = shuffle(df)
```

```python
df
```

```python
#获取数据的列名列表
columns = df.columns
columns
```

```python
#选择第一列，第二列为X的数据
X = df[columns[:2]].values
X
```

```python
#查看最后一列，标签列
df[columns[-1]]
```

```python
#选择最后一列为y
y = df[columns[-1]]
y
```

```python
#将y的值从Y,N转换为1，0
new_y = []
for item in y.values:
    if item=='Y':
        new_value = 1.0
    else:
        new_value = 0.0
    new_y.append(new_value)
y = np.array(new_y)
y
```

```python
#画出X散点图
plt.scatter(X[:, 0], X[:, 1],c=y, s=50, cmap='RdBu');
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
#使用朴素贝叶斯模型
model = GaussianNB()
#划分训练集和测试集，按0.8：0.2比例划分
train_len = int(len(X)*0.8)
X_train = X[:train_len]
X_test = X[train_len:]
y_train = y[:train_len]
y_test = y[train_len:]
#拟合模型
model.fit(X_train, y_train);
```

```python
#训练后的模型，对测试集X_test进行预测，得到预测的ypred
ypred = model.predict(X_test)
ypred
```

```python
# from sklearn.datasets import make_blobs
# X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
#画圈图
fig, ax = plt.subplots()

ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
ax.set_title('Naive Bayes Model', size=14)

xlim = (-8, 8)
ylim = (-15, 5)

xg = np.linspace(xlim[0], xlim[1], 60)
yg = np.linspace(ylim[0], ylim[1], 40)
xx, yy = np.meshgrid(xg, yg)
Xgrid = np.vstack([xx.ravel(), yy.ravel()]).T

for label, color in enumerate(['red', 'blue']):
    mask = (y == label)
    mu, std = X[mask].mean(0), X[mask].std(0)
    P = np.exp(-0.5 * (Xgrid - mu) ** 2 / std ** 2).prod(1)
    Pm = np.ma.masked_array(P, P < 0.03)
    ax.pcolorfast(xg, yg, Pm.reshape(xx.shape), alpha=0.5,
                  cmap=color.title() + 's')
    ax.contour(xx, yy, P.reshape(xx.shape),
               levels=[0.01, 0.1, 0.5, 0.9],
               colors=color, alpha=0.2)
    
ax.set(xlim=xlim, ylim=ylim)

fig.savefig('05.05-gaussian-NB.png')
# fig.show()
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
#计算准确率
acc = accuracy_score(ypred, y_test)
print(acc)
#计算混淆矩阵
mat = confusion_matrix(ypred, y_test)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=True, yticklabels=True)
plt.xlabel('true label')
plt.ylabel('predicted label');
```
